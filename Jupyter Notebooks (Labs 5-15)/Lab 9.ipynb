```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from jupyterthemes import jtplot

jtplot.style()
```

```python
def show_images(im_dict, cols, shape = (1,1), length = None, size = 8):
    if not length:
        length = len(im_dict)
    else:
        length = max(length, len(im_dict))
    w_size = shape[1] / sum(shape) * size
    h_size = shape[0] / sum(shape) * size
    rows = np.ceil(length / cols)
    fig = plt.figure(figsize=(w_size*cols, h_size*rows))
    i = 0
    for key, value in im_dict.items():
        i += 1
        sub = fig.add_subplot(rows, cols, i)
        sub.set_title(key)
        plt.axis('off')
        plt.imshow(value)
    return fig
```

```python
fn = 'image.jpg' 
img = cv.imread(fn, 0)
img_rgb = cv.cvtColor(img, cv.COLOR_GRAY2RGB)

ret,th = cv.threshold(img,20,255,cv.THRESH_BINARY)
th = cv.cvtColor(th, cv.COLOR_GRAY2RGB)
```

```python
im_dict = {'Grayscale':img_rgb, 'Binary':th}

cols = 2
size = 16
fig = show_images(im_dict, cols, shape = img.shape, size = size)
```
