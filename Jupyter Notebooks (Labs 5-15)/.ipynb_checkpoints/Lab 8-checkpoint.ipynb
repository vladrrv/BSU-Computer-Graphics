```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from jupyterthemes import jtplot

jtplot.style()
```

```python
def show_images(im_dict, cols, shape = (1,1), length = None, size = 8):
    if not length:
        length = len(im_dict)
    else:
        length = max(length, len(im_dict))
    w_size = shape[1] / sum(shape) * size
    h_size = shape[0] / sum(shape) * size
    rows = np.ceil(length / cols)
    fig = plt.figure(figsize=(w_size*cols, h_size*rows))
    i = 0
    for key, value in im_dict.items():
        i += 1
        sub = fig.add_subplot(rows, cols, i)
        sub.set_title(key)
        plt.axis('off')
        plt.imshow(value)
    return fig
```

```python
fn = 'image.jpg' 

im = cv.imread(fn);
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)

im_gray = cv.imread(fn, 0);
im_gray = cv.cvtColor(im_gray, cv.COLOR_GRAY2RGB)
```

```python
im_dict = {'Source':im, 'Grayscale':im_gray}

cols = 2
size = 16
fig = show_images(im_dict, cols, shape = im.shape, size = size)
```
